```python
for i in shipment_Controller:
    i["vol"] = i["length"] * i["width"] * i["height"]
    i["density"] = i["weight"] / i["vol"]

sorted_shipment = sorted(shipment_Controller, key=lambda x: x["density"])
tot_added = 0
```

```python
def knapSack(W, n,vol,ids):
    # Base Case
    if n == 0 or W == 0 :
        return 0
  
    # If weight of the nth item is more than Knapsack of capacity
    # W, then this item cannot be included in the optimal solution
    if (sorted_shipment[n-1]["weight"] > W  or sorted_shipment[n-1]["vol"] > vol ):
        return knapSack(W, n-1, vol)
  
    # return the maximum of two cases:
    # (1) nth item included
    # (2) not included
    else:
        return max(sorted_shipment[n-1]["density"] + knapSack(W-sorted_shipment[n-1]["weight"] , n-1, vol - sorted_shipment["vol"],ids),
                   knapSack(W, n-1, vol))
```

```python
def dp_knap(W,n,vol):
    K = [[0 for x in range(W + 1)] for x in range(n + 1)]
    # Build table K[][] in bottom up manner
    for i in range(n + 1):
        for w in range(W + 1):
            if i == 0 or w == 0:
                K[i][w] = 0
            elif sorted_shipment[n-1]["weight"] <= w and sorted_shipment[n-1]["vol"] <= vol :
                K[i][w] = max(sorted_shipment[n-1]["density"] + K[i-1][w-sorted_shipment[n-1]["weight"]],  K[i-1][w])
            else:
                K[i][w] = K[i-1][w]
  
    return K[n][W]
```

```python
def frac_knap(W,vol):
    result = []
    global rem_comp
    global tot_added
    for i in range(97, 0, -1):
        # print(sorted_shipment[i]["vol"])
        if(sorted_shipment[i]["weight"] < W and  sorted_shipment[i]["vol"] < vol and sorted_shipment[i]["weight"] < rem_comp and "added" not in sorted_shipment[i]):
            result.append(sorted_shipment[i]["id"])
            rem_comp = rem_comp - sorted_shipment[i]["weight"]
            W = W - sorted_shipment[i]["weight"]
            # print('adfads')
            sorted_shipment[i]["added"] = True
            tot_added = tot_added + 1
            vol = vol - sorted_shipment[i]["vol"]
    
    # print(result)
    return result
```
